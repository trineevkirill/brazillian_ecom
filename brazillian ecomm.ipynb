```python
# Объединяем датафреймы
geo_deliver = order_items[["order_id", "seller_id"]].merge(
    orders[["order_id", "customer_id", "order_status", "order_purchase_timestamp", "order_delivered_customer_date"]],
    how="left",
    on="order_id",
)
geo_deliver = geo_deliver.merge(sellers[["seller_id", "seller_state"]], how="left", on="seller_id")
geo_deliver = geo_deliver.merge(customers[["customer_id", "customer_state"]], how="left", on="customer_id")
geo_deliver

# Присваиваем продавцам и покупателям сведения о дистриктах
geo_deliver["seller_district"] = geo_deliver["seller_state"].map(districts)
geo_deliver["customer_district"] = geo_deliver["customer_state"].map(districts)
# Сочетание дистриктов
geo_deliver["districts"] = geo_deliver["seller_district"] + " / " + geo_deliver["customer_district"]

# Время доставки
geo_deliver["deliver_time"] = geo_deliver["order_delivered_customer_date"] - geo_deliver["order_purchase_timestamp"]

# Сортируем столбцы датафрейма
geo_deliver = geo_deliver[
    [
        "order_id",
        # 'order_item_id',
        "order_status",
        "seller_id",
        "seller_state",
        "seller_district",
        "customer_id",
        "customer_state",
        "customer_district",
        "districts",
        "deliver_time",
    ]
]

geo_deliver.head()
```

### Завершенные заказы

```python
# Фильтруем по доставленным заказам
success_delivery = geo_deliver[geo_deliver["order_status"] == "delivered"]

# Уникальные значения макроокругов для продавцов и покупателей
seller_districts = success_delivery["seller_district"].unique()
customer_districts = success_delivery["customer_district"].unique()

# Сортируем дистрикты по времени доставки по возрастанию
median_rate = success_delivery.groupby("districts")["deliver_time"].median().sort_values(ascending=True).index.tolist()

# Время доставки в часах
success_delivery["deliver_time"] = (success_delivery["deliver_time"].dt.total_seconds() / 3600).round().astype("Int64")

success_delivery.head()
```

```python
success_delivery[success_delivery.duplicated()]
```

```python
high_delivered = []
medium_delivered = []
low_delivered = []
zero_delivered = []

for district in median_rate:
    temp_df = success_delivery[success_delivery["districts"] == district]
    if temp_df.shape[0] == 0:
        zero_delivered.append(district)
    elif temp_df.shape[0] < 100:
        low_delivered.append(district)
    elif temp_df.shape[0] < 500:
        medium_delivered.append(district)
    else:
        high_delivered.append(district)
```

```python
custom_palette = {
    district: "#00c137" if district in high_delivered else ("#1f65b4" if district in low_delivered else "#e0e0e0")
    for district in median_rate
}

plt.figure(figsize=(16, 10))

# median_rate.index — это список районов, отсортированный по возрастанию медианы
sns.boxplot(
    data=success_delivery, x="deliver_time", y="districts", order=median_rate, hue="districts", palette=custom_palette
)

# Исправляем названия на соответствующие вашим данным
plt.title('Распределение времени доставки по районам (в часах) по направлению "Продавец / Покупатель"', fontsize=14)
plt.xlabel("Время доставки (в часах)", fontsize=12)
plt.ylabel("Районы", fontsize=12)
plt.grid(axis="x", alpha=0.5)
plt.grid(axis="y", alpha=0.2)

plt.show()
```

Необходимо унифицировать все размеры посылок по четкой градации. По аналогии с размерами посылок Почты России:

**1. Размер**

| Размер | Описание | Макс. сумма трех сторон (Д + Ш + В) | Макс. длина стороны
| :-----: | :----- | :-----: | :-----: | 
| **XS** | Мелкие посылки (документы, чехлы, бижутерия) | до 35 см | до 20 см | 
| **S** | Маленькие посылки (книги, косметика, легкая одежда) | до 60 см | до 35 см | 
| **M** | Средние (обувные коробки, мелкая бытовая техника, посуда) | до 90 см | до 50 см | 
| **L** | Большие (верхняя одежда, средняя бытовая техника) | до 120 см | до 80 см | 
| **XL** | Очень большие (чемоданы, автозапчасти, детские коляски) | до 150 см | до 120 см | 
| **XXL** | Все, что превышает лимиты **XL** | более 150 см | более 120 см | 


**2. Вес**

**Легковесные (до 2 кг):** Мелкие пакеты, письма, одежда. С ними легко справляется один человек, можно нести в руках.

**Стандартные (от 2 до 10 кг):** Обычные посылки, бытовая техника, инструмент. Стандартный лимит для большинства ПВЗ (пунктов выдачи).

**Тяжелые (от 10 до 30 кг):** Крупные коробки, мебель в разобранном виде. Требуют осторожности при ручном переносе.

**Сверхтяжелые / Крупногабаритные (более 30 кг):** По стандартам большинства служб доставки, всё, что весит более 30-31.5 кг, официально считается тяжеловесным (крупногабаритным) грузом. Для таких посылок часто требуются два грузчика, гидравлическая тележка (рохля) или курьер на грузовом автомобиле.


```python
# Размер

long_delivery_df['sum_side'] = long_delivery_df['product_length_cm'] + long_delivery_df['product_height_cm'] + long_delivery_df['product_width_cm']

length_bins = [0, 35, 60, 90, 120, 150, np.inf]
length_labels = ["XS", "S", "M", "L", "XL", "XXL"]
long_delivery_df['length'] = pd.cut(long_delivery_df['sum_side'], bins=length_bins, labels=length_labels, right=False)

# Вес

weight_bins = [0, 2000, 10000, 30000, np.inf]
weight_labels = ["Легкие", "Средние", "Тяжелые", "Сверхтяжелые"]
long_delivery_df["weight"] = pd.cut(long_delivery_df["product_weight_g"], bins=weight_bins, labels=weight_labels, right=False)

long_delivery_df.head()
```

```python
long_delivery_df_grouped = long_delivery_df.groupby(['length', 'weight'])['order_id'].nunique()
long_delivery_df_grouped = long_delivery_df_grouped.reset_index()
long_delivery_df_grouped
```
