```python
# Объединяем датафреймы
geo_deliver = order_items[["order_id", "seller_id"]].merge(
    orders[["order_id", "customer_id", "order_status", "order_purchase_timestamp", "order_delivered_customer_date"]],
    how="left",
    on="order_id",
)
geo_deliver = geo_deliver.merge(sellers[["seller_id", "seller_state"]], how="left", on="seller_id")
geo_deliver = geo_deliver.merge(customers[["customer_id", "customer_state"]], how="left", on="customer_id")

# Присваиваем продавцам и покупателям сведения о дистриктах
geo_deliver["seller_district"] = geo_deliver["seller_state"].map(districts)
geo_deliver["customer_district"] = geo_deliver["customer_state"].map(districts)
# Сочетание дистриктов
geo_deliver["districts"] = geo_deliver["customer_district"] + " / " + geo_deliver["seller_district"]

# Время доставки
geo_deliver["deliver_time"] = geo_deliver["order_delivered_customer_date"] - geo_deliver["order_purchase_timestamp"]

# Сортируем столбцы датафрейма
geo_deliver = geo_deliver[
    [
        "order_id",
        "order_status",
        "seller_id",
        "seller_state",
        "seller_district",
        "customer_id",
        "customer_state",
        "customer_district",
        "districts",
        "deliver_time",
    ]
]

geo_deliver.head()
```

### Завершенные заказы

```python
# Фильтруем по доставленным заказам
success_delivery = geo_deliver[geo_deliver["order_status"] == "delivered"]

# Уникальные значения макроокругов для продавцов и покупателей
seller_districts = success_delivery["seller_district"].unique()
customer_districts = success_delivery["customer_district"].unique()

# Сортируем дистрикты по времени доставки по возрастанию
median_rate = success_delivery.groupby("districts")["deliver_time"].median().sort_values(ascending=True).index.tolist()

# Время доставки в часах
success_delivery["deliver_time"] = (success_delivery["deliver_time"].dt.total_seconds() / 3600).round().astype("Int64")

success_delivery.head()
```

```python
high_delivered = []
medium_delivered = []
low_delivered = []
zero_delivered = []

for district in median_rate:
    temp_df = success_delivery[success_delivery['districts']==district]
    if temp_df.shape[0]==0:
        zero_delivered.append(district)
    elif temp_df.shape[0]<100:
        low_delivered.append(district)
    elif temp_df.shape[0]<1000:
        medium_delivered.append(district)
    else:
        high_delivered.append(district)
```

```python
zero_delivered
```

```python
custom_palette = {
    district: "#00c137" if district in high_delivered else (
              "#1f65b4" if district in low_delivered else "#e0e0e0") 
    for district in median_rate
}

plt.figure(figsize=(16, 10))

# median_rate.index — это список районов, отсортированный по возрастанию медианы
sns.boxplot(
    data=success_delivery, x="deliver_time", y="districts", order=median_rate, hue="districts", palette=custom_palette
)

# Исправляем названия на соответствующие вашим данным
plt.title('Распределение времени доставки по районам (в часах) по направлению "Продавец / Покупатель"', fontsize=14)
plt.xlabel("Время доставки (в часах)", fontsize=12)
plt.ylabel("Районы", fontsize=12)
plt.grid(axis="x", alpha=0.5)
plt.grid(axis="y", alpha=0.2)

plt.show()
```

### Приближение

```python
for dist in low_delivered:
    temp_df = success_delivery[success_delivery["districts"] == dist]
    min = pd.to_timedelta(temp_df["deliver_time"].min(), unit="D").floor('s')
    quant_25 = pd.to_timedelta(temp_df["deliver_time"].quantile(0.25), unit="D").floor('s')
    quant_50 = pd.to_timedelta(temp_df["deliver_time"].median(), unit="D").floor('s')
    quant_75 = pd.to_timedelta(temp_df["deliver_time"].quantile(0.75), unit="D").floor('s')
    max = pd.to_timedelta(temp_df["deliver_time"].max(), unit="D").floor('s')
    std = pd.to_timedelta(temp_df["deliver_time"].std(), unit="D").floor('s')

    # Сведения
    print(f"{'-' * 10} Дистрикт: {dist} {'-' * 10}")
    print(f'- Количество заказов \t\t{temp_df.shape[0]}')
    print(f"- Минимум: \t\t\t{min/60}")
    print(f"- 25 процентиль: \t\t{quant_25/60}")
    print(f"- 50 процентиль: \t\t{quant_50/60}")
    print(f"- 75 процентиль: \t\t{quant_75/60}")
    print(f"- Максимум: \t\t\t{max/60}")
    print(f"- Стандартное отклонение: \t{std/60}")
    print("")
```
